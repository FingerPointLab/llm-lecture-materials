```python
# 문서 출력 도우미 함수
def pretty_print_docs(docs):
    print(
        f"\n{'-' * 100}\n".join(
            [f"Document {i+1}:\n\n" + d.page_content for i, d in enumerate(docs)]
        )
    )
```

```python
import base64
import requests

# Load the text file
path = "https://raw.githubusercontent.com/FingerPointLab/sds-lecture/refs/heads/main/%EC%8B%A4%EC%8A%B5%EB%8D%B0%EC%9D%B4%ED%84%B0/%EC%95%BD%ED%95%99%EC%A0%95%EB%B3%B4%EC%9B%90-%EC%95%BD%EB%AC%BC%EB%B0%B1%EA%B3%BC%26%EB%B6%80%EC%9E%91%EC%9A%A9%EB%A6%AC%ED%8F%AC%ED%8A%B8.txt"

req = requests.get(path)
req_text = req.text
print(req_text)
```

```python
from langchain.docstore.document import Document

documents = [Document(page_content=req_text)]
documents
```

### 문서 변환 및 저장하기(임베딩, 벡터스토어)

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

# # 문서 로드
# documents = TextLoader("https://raw.githubusercontent.com/FingerPointLab/sds-lecture/refs/heads/main/%EC%8B%A4%EC%8A%B5%EB%8D%B0%EC%9D%B4%ED%84%B0/%EC%95%BD%ED%95%99%EC%A0%95%EB%B3%B4%EC%9B%90-%EC%95%BD%EB%AC%BC%EB%B0%B1%EA%B3%BC%26%EB%B6%80%EC%9E%91%EC%9A%A9%EB%A6%AC%ED%8F%AC%ED%8A%B8.txt").load()

# 텍스트 분할기 설정
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)

# 문서 분할
texts = text_splitter.split_documents(documents)

# 임베딩 모델 설정
embeddingsModel = HuggingFaceEmbeddings(
    model_name="sentence-transformers/msmarco-distilbert-dot-v5"
)

# 문서로부터 FAISS 인덱스 생성 및 검색기 설정
retriever = FAISS.from_documents(texts, embeddingsModel).as_retriever(
    search_kwargs={"k": 10}
)

# 질의 설정
query = "불면증 개선이 필요할 때"

# 질의 수행 및 결과 문서 반환
docs = retriever.invoke(query)

# 결과 문서 출력
pretty_print_docs(docs)
```

### CrossEncoderReranker

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import CrossEncoderReranker
from langchain_community.cross_encoders import HuggingFaceCrossEncoder

# 모델 초기화
model = HuggingFaceCrossEncoder(model_name="BAAI/bge-reranker-v2-m3")

# 상위 3개의 문서 선택
compressor = CrossEncoderReranker(model=model, top_n=3)

# 문서 압축 검색기 초기화
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=retriever
)

# 질의 설정
query = "불면증 개선이 필요할 때"

# 압축된 문서 검색
compressed_docs = compression_retriever.invoke(query)

# 문서 출력
pretty_print_docs(compressed_docs)
```
